Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection & Pre-Processing

```python
#loading data from csv file to pandas DataFrame
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
# replace the null values with a null string
mail_data =raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
#printing first five rows
mail_data.head()
```

```python
#checking th number of rows AND COLUMNS
mail_data.shape
```

Label Encoding

```python
#label spam mail as 0 , ham mail as 1

mail_data.loc[mail_data['Category'] == 'spam','Category',]=0
mail_data.loc[mail_data['Category']=='ham','Category',]=1
```

```python
#seperating the data set as texts and label

x = mail_data['Message']
y = mail_data['Category']
```

```python
print(x)
```

```python
print(y)
```

Splitting data into training data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

Feature Extraction

```python
#transform the text data to feature vectors that can be used as input to the logistic regression

feature_extraction=TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

#convert y_train and y_test values as integers

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(y_train)
```

```python
print(x_train)
```

```python
print(x_train_features)
```

Training the model

Logistic regression

```python
model = LogisticRegression()
```

```python
#Training the logistic regression model with training data
model.fit(x_train_features,y_train)
```

Evaluating the trained model

```python
#prediction on taining data

prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train,prediction_on_training_data)
```

```python
print('Accuracy on traing data : ',accuracy_on_training_data)
```

```python
#prediction on test data

prediction_on_test_data = model.predict(x_test_features)
accuracy_on_test_data = accuracy_score(y_test,prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```

Building a predictive system

```python
input_mail=["i have a work"]
#convert text into feature vecotors
input_data_features = feature_extraction.transform(input_mail)

#making prediction

prediction = model.predict(input_data_features)
print(prediction)

if (prediction[0]==1):
  print('Ham mail')
else:
  print('Spam mail')
```
